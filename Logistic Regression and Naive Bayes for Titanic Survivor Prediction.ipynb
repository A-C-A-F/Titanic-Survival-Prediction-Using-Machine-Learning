```python
# Let's drop the embarked, Ticket, passengerID, and Name as well to simplify our analysis as we wouldn't need these info.
titanic_df.drop(['Embarked', 'Ticket', 'PassengerId', 'Name'], axis=1, inplace=True)
```

```python
titanic_df
```

```python
# Let's view the data one more time!
sns.heatmap(titanic_df.isnull(), yticklabels=False, cbar=False, cmap='Blues')
```

There seems alot of missing data in Age column. We would not want to delete this as we can still gain insights from this. The solution we can do about this is by using some art. Getting the average age for eeach sex type and populate the missing value with it repespective to its gender.

```python
# Let's get the average age for male and female
plt.figure(figsize=(15,10))
sns.boxplot(x = 'Sex', y = 'Age', data = titanic_df)
```

```python
# Based on the plot above, the average age for male is (~29) and for female is (~25).
# "~" means approximately.
```

```python
def Fill_Ages(data):
    age = data[0]
    sex = data[1]
    
    if pd.isnull(age):
        if sex == 'male':
            return 29
        else:
            return 25
    else:
        return age
```

```python
titanic_df['Age'] = titanic_df[['Age', 'Sex']].apply(Fill_Ages, axis = 1)
```

```python
# Let's view the data one more time!
sns.heatmap(titanic_df.isnull(), yticklabels = False, cbar = False, cmap="Blues")
```

```python
titanic_df
```

```python
# You just need one column only to represent male or female
pd.get_dummies(titanic_df['Sex'])
```

```python
male = pd.get_dummies(titanic_df['Sex'], drop_first = True)
```

```python
male
```

```python
# first let's drop the embarked and sex 
titanic_df.drop(['Sex'], axis = 1, inplace=True)
```

```python
titanic_df
```

```python
# Now let's add the encoded column male again
titanic_df = pd.concat([titanic_df, male], axis=1)
```

```python
# Let's rename the 'male' column to Sex
titanic_df.rename(columns={'male': 'Sex [M(1)/F(0)]'}, inplace=True)
```

# TASK #6: TRAIN LOGISTIC REGRESSION CLASSIFIER MODEL

```python
#Let's drop the target coloumn before we do train test split
X = titanic_df.drop('Survived', axis = 1).values
y = titanic_df['Survived'].values
```

So whenever we train any machine learning model, what we do
generally is we use the bulk of the data, maybe
80% for example of the data to train our model and then once
the model is trained, what we do is we use the remaining data
set, which is let's say 20% of the data as a testing dataset
to assess the performance of the model on a new data set that
the model has never seen before during training.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# Fitting Logistic Regression to the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)
```

# TASK #7: ASSESS TRAINED MODEL PERFORMANCE

```python
y_predict_test = classifier.predict(X_test)
y_predict_test
```

. And what I could do right now is I wanted to compare

in the real world.
What actually happened in reality.
And in order to do that, I wanted to print or plot what

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot = True, fmt = "d")
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict_test))
```
